## Import Required Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## Define Helper Functions

```python
def sample_image(image, factor):
    height, width = image.shape[:2]
    sampled_image = cv2.resize(image, (width // factor, height // factor), interpolation=cv2.INTER_NEAREST)
    return sampled_image

def quantize_image(image, levels):
    quantized_image = np.floor(image / (256 // levels)) * (256 // levels)
    quantized_image = quantized_image.astype(np.uint8)
    return quantized_image

def plot_images(original, sampled, quantized, sampling_factor, quantization_levels):
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(sampled, cmap='gray')
    plt.title(f'Sampled Image (Factor={sampling_factor})')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(quantized, cmap='gray')
    plt.title(f'Quantized Image (Levels={quantization_levels})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
```

## Load Original Image

```python
image_path = 'images/lena_gray_256.tif'
original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if original_image is None:
    print(f"Error: Unable to load image at {image_path}")
```

# Task #2: Image Operations

Perform various arithmetic and logical operations on images.

## Load and Resize Two Images

```python
img1_path = 'images/lena_gray_256.tif'
img2_path = 'images/cameraman.tif'

img1 = Image.open(img1_path)
img2 = Image.open(img2_path)

resize = (400, 400)
img1 = img1.resize(resize, Image.Resampling.LANCZOS)
img2 = img2.resize(resize, Image.Resampling.LANCZOS)

im1arr = np.asarray(img1)
im2arr = np.asarray(img2)
```

## Operation 1: Subtract Two Images

```python
subtraction = cv2.subtract(im1arr, im2arr)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(im1arr, cmap='gray')
plt.title('Image 1')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(im2arr, cmap='gray')
plt.title('Image 2')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(subtraction, cmap='gray')
plt.title('Subtraction (Image1 - Image2)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Operation 2: Add Constant Value of 175 to Image

```python
addition_constant = cv2.add(im1arr, np.full(im1arr.shape, 175, dtype=np.uint8))

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(im1arr, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(addition_constant, cmap='gray')
plt.title('Image + 175')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Operation 3: Set Difference (A - B)

```python
set_difference = cv2.subtract(im1arr, cv2.bitwise_and(im1arr, im2arr))

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(im1arr, cmap='gray')
plt.title('Image 1 (A)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(im2arr, cmap='gray')
plt.title('Image 2 (B)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(set_difference, cmap='gray')
plt.title('Set Difference (A - B)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Operation 4: Symmetric Difference (A XOR B)

```python
symmetric_diff = cv2.bitwise_xor(im1arr, im2arr)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(im1arr, cmap='gray')
plt.title('Image 1 (A)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(im2arr, cmap='gray')
plt.title('Image 2 (B)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(symmetric_diff, cmap='gray')
plt.title('Symmetric Difference (A XOR B)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Operation 5: Intersection (A AND B)

```python
intersection = cv2.bitwise_and(im1arr, im2arr)

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(im1arr, cmap='gray')
plt.title('Image 1 (A)')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(im2arr, cmap='gray')
plt.title('Image 2 (B)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(intersection, cmap='gray')
plt.title('Intersection (A AND B)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Procedural Example – Unsharp Masking

```python
# Read the image
image = cv2.imread('boat.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR to RGB

# Gaussian sigma
sigma = 2

# parameter
amount = 1.5

# Convert image to float32 for processing
image_float = image.astype(np.float32) / 255.0

# Apply Gaussian blur
blurred = cv2.GaussianBlur(image_float, (0, 0), sigmaX=sigma, sigmaY=sigma)

# Perform Unsharp Masking
sharpened = np.clip(image_float + amount * (image_float - blurred), 0, 1)

# Plot original, blurred, and sharpened images
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(blurred)
ax[1].set_title("Blurred Image")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened (Unsharp Masking)")
ax[2].axis("off")

plt.suptitle("Procedural Example: Unsharp Masking", fontsize=14)
plt.tight_layout()
plt.savefig('unsharp_masking.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Assessment Task #1 – 7×7 Box Filter (Smoothing)

```python
# Read the image
image = cv2.imread('boat.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply 7x7 box filter (normalized box filter = averaging filter)
kernel_size = 7
box_filtered = cv2.blur(image, (kernel_size, kernel_size))

# Display original and smoothed image side by side
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(box_filtered)
ax[1].set_title("Smoothed Image (7×7 Box Filter)")
ax[1].axis("off")

plt.suptitle("Task #1: 7×7 Box Filter", fontsize=14)
plt.tight_layout()
plt.savefig('task1_box_filter.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Kernel size: {kernel_size}×{kernel_size}")
print("Effect: Uniform averaging of pixel neighborhoods — reduces noise but also blurs edges.")
```

## Assessment Task #2 – 5×5 and 21×21 Gaussian Filters

```python
# Read the image
image = cv2.imread('boat.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply 5x5 Gaussian filter
gauss_5x5 = cv2.GaussianBlur(image, (5, 5), 0)

# Apply 21x21 Gaussian filter
gauss_21x21 = cv2.GaussianBlur(image, (21, 21), 0)

# Display all three images side by side
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gauss_5x5)
ax[1].set_title("5×5 Gaussian Filter")
ax[1].axis("off")

ax[2].imshow(gauss_21x21)
ax[2].set_title("21×21 Gaussian Filter")
ax[2].axis("off")

plt.suptitle("Task #2: Gaussian Filters – Effect of Kernel Size", fontsize=14)
plt.tight_layout()
plt.savefig('task2_gaussian_filters.png', dpi=150, bbox_inches='tight')
plt.show()

print("Observation: Larger kernel size (21×21) produces stronger blurring compared to smaller kernel (5×5).")
```

## Assessment Task #3 – Laplacian Filter for Sharpening

```python
# Step 1: Read the image
image = cv2.imread('boat.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Step 2: Convert image to float32
image_float = image.astype(np.float32) / 255.0

# Step 3: Use the cv2.Laplacian() function (ksize=3 → 3×3 kernel)
laplacian = cv2.Laplacian(image_float, cv2.CV_32F, ksize=3)

# Step 4: Sharpen the image
sharpened = np.clip(image_float - laplacian, 0, 1)

# Step 5: Plot the original, Laplacian output, and sharpened image
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

# Normalize Laplacian for display
lap_display = cv2.normalize(np.abs(laplacian), None, 0, 1, cv2.NORM_MINMAX)
ax[1].imshow(lap_display)
ax[1].set_title("Laplacian Filter Output (3×3)")
ax[1].axis("off")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened Image")
ax[2].axis("off")

plt.suptitle("Task #3: Laplacian Sharpening", fontsize=14)
plt.tight_layout()
plt.savefig('task3_laplacian_sharpening.png', dpi=150, bbox_inches='tight')
plt.show()

print("Observation: The Laplacian filter highlights edges/fine details.")
print("Subtracting it from the original amplifies those details, resulting in a sharper image.")
```
